HANGMAN OYUNU

```python
import random

def play_hangman():
  words= ["python","hangman","veri","yapayzeka","teknofest"]
  word = random.choice(words)
  guessed_letters=[]
  attempts=0
  max_attempts=4
  done = False

  print("Hnagman Oyununa Hosgeldiniz!")

  while not done:
        guessed_word = ""
        for letter in word:
            if letter in guessed_letters:
                guessed_word += letter
            else:
                guessed_word += "_"

        print("Kelime: ", " ".join(guessed_word))


        guess=input("Guess a letter: ").lower()
        if guess in guessed_letters:
          print("You already guessed that letter!")
        elif guess in word:
          print("correct!")
          guessed_letters.append(guess)
        else:
          print("Incorrect")
          attempts+=1
          guessed_letters.append(guess)

        if attempts>=max_attempts:
          done=True
          print("You lost! The word was" , word)
        elif set(word)==set(guessed_letters):
          done=True
          print("You won! The word was" , word)

play_hangman()





```

SAYI TAHMİNİ OYUNU


```python
import random

def NumberGuessing_Game():

   target_number=random.randint(1,25)
   attempts=0

   print("Sayı Tahmini Oyununa Hoşgeldiniz!")
   print("1 ile 25 arasında bir sayı tahmin ediniz.")

   while True:
    guess=int(input("Tahmininizi Giriniz:"))
    attempts+=1
    if guess < target_number:
      print("Daha büyük bir sayı giriniz")
    elif guess > target_number:
      print("Daha küçük bir sayı giriniz")

    else:
      print("TEBRİKLER...Doğru Tahmin",attempts,"denemede bildiniz")


NumberGuessing_Game()
```

```python
def check_password_strength(password):
    # Parola uzunluğu kontrolü
    if len(password) < 8:
        return "Parola çok kısa, en az 8 karakter olmalıdır."

    # Kriterler için başlangıç puanı
    score = 0

    # Büyük harf kontrolü
    if any(char.isupper() for char in password):
        score += 1

    # Küçük harf kontrolü
    if any(char.islower() for char in password):
        score += 1

    # Rakam kontrolü
    if any(char.isdigit() for char in password):
        score += 1

    # Özel karakter kontrolü
    special_characters = "!@#$%^&*()-_+=<>?/{}[]|\\,."
    if any(char in special_characters for char in password):
        score += 1

    # Puanlamaya göre güç seviyesi belirleme
    if score == 1:
        return "Zayıf"
    elif score == 2:
        return "Orta"
    elif score == 3:
        return "İyi"
    elif score == 4:
        return "Çok İyi"

def main():
    password = input("Parolanızı girin: ")
    strength = check_password_strength(password)
    print("Parolanızın gücü: Güçlü")

if __name__ == "__main__":
    main()
```

```python
def check_password_strength(password):
    # Parola uzunluğu kontrolü
    if len(password) < 8:
        return "Parola çok kısa, en az 8 karakter olmalıdır."

    # Karakter türü kümesi tanımlamaları
    uppercase_set = set("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    lowercase_set = set("abcdefghijklmnopqrstuvwxyz")
    digit_set = set("0123456789")
    special_characters_set = set("!@#$%^&*()-_+=<>?/{}[]|\\,.")

    # Karakter türü varlığını kontrol eden bool değişkenler
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    # Parola üzerinde döngülerle karakter türü varlığını kontrol etme
    for char in password:
        if char in uppercase_set:
            has_upper = True
        elif char in lowercase_set:
            has_lower = True
        elif char in digit_set:
            has_digit = True
        elif char in special_characters_set:
            has_special = True

    # Puanlamaya göre güç seviyesi belirleme
    score = sum([has_upper, has_lower, has_digit, has_special])
    if score == 1:
        return "Zayıf"
    elif score == 2:
        return "Orta"
    elif score == 3:
        return "İyi"
    elif score == 4:
        return "Çok İyi"

def main():
    password = input("Parolanızı girin: ")
    strength = check_password_strength(password)
    print("Parolanızın gücü:",strength)

# main() fonksiyonunu doğrudan çağırarak programın başlamasını sağlıyoruz
if __name__ == "__main__":
    main()
```
